```python
x = np.linspace(0, 1, 101)

Y = np.sin(2 * np.pi * (x[None, :] + np.linspace(0, 1, 7)[:, None]) ) 
# Example categorigal
styler.set_color_palette('sequential_teal')



# Example 2D

fig, axes = plt.subplots(1, 3, figsize = styler.figsize(1, 3))
axes[0].plot(x, Y.T)
```
